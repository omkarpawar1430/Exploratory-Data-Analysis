### Changing Values:

```python
df_copy['Size'].head(10)
```

```python
df_copy['Size'].dtype
```

```python
# df_copy['Size'].unique() # gives unique values 
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('M', '000')
df_copy['Size'] = df_copy['Size'].str.replace('k', '')
df_copy['Size'].head(10)
```

```python
df_copy['Size'].unique() # 'Varies with device'??? 
```

```python
import numpy as np 
df_copy['Size'] = df_copy['Size'].str.replace('Varies with device', str(np.nan))
df_copy['Size'].unique()
```

```python
df_copy['Size'] = df_copy['Size'].astype('float')
df_copy['Size'].dtype
```

Values >= 10 are getting converted to KB in right way... 

But
Values less than 10 are not converted to KB in right way... 

'8.7000' - wrong |
'8700.0' - correct

```python
8.7 * 1000 # we can do this simple mathematical operation on every value less than 10 
```

```python
for i in df_copy['Size']:
    if i < 10:
        df_copy['Size'] = df_copy['Size'].replace(i, i*1000)
```

```python
df_copy['Size'].head() # dtype: float64
```

```python
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```
