### Central Limit Theorem:

It states that as the sample size of a random variable increases, the distribution of the sample means approaches a normal distribution, regardless of the distribution of the population from which the samples are drawn.

```python
population = np.random.binomial(10, 0.5, 10000)
population
```

```python
len(population)
```

```python
import seaborn as sns
```

```python
sns.distplot(population)
```

```python
len(population)
```

```python
sample_size =  int(len(population)*0.30)
```

```python
sample_mean = []
for i in range(1, 31):
    # print(np.random.choice(population,size= sample_size))
    sample = np.random.choice(population,size= sample_size)
    sample_mean.append(np.mean(sample))
sns.distplot(sample_mean)
```

 result is approximately normal distribution, hence central limit theorem is proved.

```python
sample_mean = []
for i in range(1, 100):
    # print(np.random.choice(population,size= sample_size))
    sample = np.random.choice(population,size= sample_size)
    sample_mean.append(np.mean(sample))
sns.distplot(sample_mean)
```

```python
# increasing the number of samples: 

sample_mean = []
for i in range(1, 500):
    # print(np.random.choice(population,size= sample_size))
    sample = np.random.choice(population,size= sample_size)
    sample_mean.append(np.mean(sample))
sns.distplot(sample_mean)
```

```python
# increasing the number of samples:

sample_mean = []
for i in range(1, 5000):
    # print(np.random.choice(population,size= sample_size))
    sample = np.random.choice(population,size= sample_size)
    sample_mean.append(np.mean(sample))
sns.distplot(sample_mean)
```

As we increase the sample size our distribution becomes more close to normal distribution.
